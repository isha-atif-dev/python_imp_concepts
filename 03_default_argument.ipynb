# Arguments 

These are the actual values that are passed to the function when calling the function.

Note : parameter(names used in function definition)


```python
def sum1(num1, num2):   # parameters
    return num1 + num2


sum1(2,3)   #arguments 
```

# Types of arguments

1. Postional arguments

2. keyword arguments 

3. default arguments 

4. Variable length arguments

   1. *args
   2. *kwargs


```python
# positional

def sum2(num1, num2):
    return num1 + num2


sum2(2,3)    #these are postional arguments here ORDER MATTERS cuz num1 = 2 and num2 = 3
```

```python
# keyword 


def sum3(num1 , num2):
    return num1 + num2 

sum3(num1 = 3, num2 = 2)  # we specify the parameters names here so order dont matter 
```

```python
# default 

def sum4(num1 = 4, num2 = 3):
    return num1 + num2

sum4(3,)   #  id we dont give any arguments here then the default arguments will be used 
```

```python
# *args
#  allows the function to take any number of positional arguments 

def add_all(*nums):
    return sum(nums)

print(add_all(1,2,3,4))   #it actually becomes tuple which is immutable
```

```python
# **kwargs 
# keyword args means it allows any number of keywords arguments 

def my_addition(**info):
    return info

my_addition(name = "isha", age = 23)

# it becomes dictionary
```

# Now mutable and immutable objects work inside functions 

# Case 1 

- Mutable object modified


```python
def add_item(list):
    return list.append(3)

list1 = [1,2]
add_item(list1)
print(list1)    # here it modified the list (original object) in place 
```

# Case 2

- Here the reassigment trap with the mutable objects 

```python
def add_Items(list2):
    new_list = list2 + [3]
    return new_list

list3 = [1,2]

print(add_Items(list3))
print(list3)

# this shows that it dont modify the original it creates a new object
```

# CASE 3

- Immutable objects 


```python
def add_one(num):
    num += 1

a = 4
add_one(a)
print(a)

# shows immutable objects cannot be changed 
```

# Default argument trap



<!-- the problem is using the mutable object as a default argument -->

Note : the default values like "def sum(num = 2)" is evaluated once whne the functions is defined meaning it dont create again and again when the function is call

here in this example below:


```python
def f(x, cache = []):
    cache.append(x)
    return cache

# here python create one list at the time of definition which will be reused for every function call

```

- What Actually Happens in Memory

When Python reads the function definition:

It does this internally:

Create empty list object → []
Store it as default value for parameter cache

Now every time you call:

f(1)
f(2)

Both calls use the SAME list.

So:

First call:

[] → append 1 → [1]

Second call:

[1] → append 2 → [1, 2]

It keeps growing.

4️⃣ Why Is This Dangerous?

Because the function now has hidden shared state.

That means:

The output depends on previous calls.

That is very dangerous in:

ML pipelines

APIs

Production systems

Because behavior becomes unpredictable.


# The Correct Pattern


```python
def f(x, cache=None):
    if cache is None:
        cache = []
    cache.append(x)
    return cache

# Now:

# Each call gets a fresh list.

# No hidden shared state.
```
