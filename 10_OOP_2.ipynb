# Instance Variables vs Class Variables


# Instance variables

An instance variable belongs to a specific object (instance).

- each object has its own value

They are usually created inside __init__.

for example

```python
class Bank():

    def __init__(self,balance):
        self.balance = balance

    def deposit(self,amount):
        self.balance += amount

acc1 = Bank(1000)
acc2 = Bank(2000)

# so here each object have balance variable and each having their own balance
# so balance is instance variable
```

# Class Variables

A class variablec belong to itself not to individual objects

like every object have the same thing 

for example

```python
class Bank_2:

    bank_name = 'HSBC'

    def __init__(self,balance):
        self.balance = balance



acc1 = Bank_2(1000)
acc2 = Bank_2(2000)
print(acc1.bank_name)
print(acc2.bank_name)

# but if 

acc1.bank_name = "abc"
print(acc1.bank_name)

# here the instance var takes over the class variables

# and its called Shadowing
```

# Encapsulation.

- Its the concept of bundling data and the methods that operate on that data inside a single unit(object)

Object = Data + Functions that work on that data

without it 
- the data will be global and its not specific to each object and its hard to maintain that code

for example:

```python
balance = 1000

def deposit(amount):
    global balance
    balance += amount

# this is the bad design
```

- With encapsulation (OOP design)

```python
# We group data and behavior together.

class BankAccount:

    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

# Now the object contains everything.

acc = BankAccount(1000)
acc.deposit(200)
```

# Why encapsulation is useful

- Organization: Everything related to the object is in one place.
