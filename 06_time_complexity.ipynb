# What is Time complexity

Time complexity describes how the number of operations an algorithm performs grows as the input size grows.

If input size becomes 10× bigger, how does runtime grow?

- Why we use it ?

we use it to compare algorithms and choose the one that scales better for large data 

# What is Big-O notation?

Big-O notation is a way to describe the time complexity of an algorithm.

- It tells us:

How the number of operations grows as the input size grows, focusing on the worst-case growth.


# how to write it?

O(...)

O means the "Order of growth"

O(n)  means the number of operations grows proportionally with n(input size )

- Big-O ignores constants and keeps only the dominant growth pattern.

# Examples with all the Big-O types
# O(1) — Constant Time

```python
# Example 1 : direct memory access.

nums = [1,2,3,4]
print(nums[2]) #python directly jumps to index 2 so just one oepration

# now even if the list has 1000 elements it doesnt matter the complexity remain the same which is constant O(1)
```

```python
# Example 2 : dict lookup
dict_1  = {
    "name" : "isha",
    "age" : 23
}
print(dict_1["name"])    # dict used hash table values to locate and object in the memory

# Steps:

# 1.Python computes hash("name")
# 2. That hash tells Python where the key is stored
# 3. Python jumps directly to that slot

# So lookup takes constant time on average.

# O(1)

```

```python
# Example 3 :set.add(x) — Adding to a Set

set_1 = {1,2,3}
set_1.add(4)
set_1

# same as dict set also uses hash table to locate directly so it also takes constant time on avg
```

# O(n) (linear time)
- The number of operations grows directly with the input size n.

so if the input doubles the operations double

```python
# Example 1 :
nums = [1,2,3,4,5]

for n in nums:
    print(n)

    # here must go through every element in the list so number of operations grow linearly with n
    # O(n)
```

```python
# Example 2 : Finding Maximum

nums = [5,6,3,2,7,4,10,1,5,4]

max_value = nums[0]
for num in nums:
    if num > max_value:
        max_value = num

max_value   # here also python check every element and if input size grows ha operations grows propotionally
# O(n)
```

```python
# Example 3 : Membership in List
# first we look member ship in set now we look it in list 

nums = [1,2,3,4,5,6]
print(3 in nums)

# O(n)
```

# O(n²) — Quadratic Time

- The number of operations grows proportional to n × n (n squared).

This usually happens when you loop over the same data inside another loop.

```python
#Simple example
nums = [1,2,3,4,5,6,7]

for i in nums:
    for j in nums:
        print(i ,j)

# Total operations: 
# 7 × 7 = 49

#so  O(n²)
```

```python
# Example 1: Finding duplicated using nested loops

nums = [1,2,3,4,5,3,6,7,2,8]

for i in range(len(nums)):
    for j in range(i+1,len(nums)):
        if nums[i] == nums[j]:
            print(nums[i])
```

# O(log n) — Logarithmic Time

binary search (O(log n)) works only if the list is already sorted.

- O(log n) happens when the problem size is cut in half each step.

O(log n) usually appears when an algorithm repeatedly halves the search space, and sorted data is one way that makes this possible.

```python
# Example — Searching in a Sorted List

nums = [1, 3, 5, 7, 9, 11, 13, 15]
target = 9


def binary_search(nums,target):
    left = 0
    right = len(nums) - 1
    while left <= right:
        mid = (left + right)// 2
        if nums[mid] == target:
            return True
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return False

binary_search(nums,target)   
```

[1,3,5,7,9,11,13]  size = 7

[9,11,13]          size = 3

[9]                size = 1

Each step roughly halves the remaining data.

That’s why the complexity is:

O(log n)

# IMP : So log n = number of times you divide by 2 until reaching 1.

# O(n log n)

- O(n log n) describes an algorithm where the total number of operations grows proportional to n × log n as the input size increases.

In simple terms:

The algorithm processes n elements, and this work happens across about log n steps or levels.

Example where it appears

Common examples:

Merge Sort

Quick Sort (average case)

Heap Sort

Python's sorted() / .sort()

These algorithms repeatedly divide data and then process all elements while combining results, which produces the n log n growth pattern.

```python
nums = [14,7,3,12]

def merge_sort(nums):
    if len(nums) <= 1:
        return nums
    
    mid = len(nums) // 2
    left = merge_sort(nums[:mid])
    right = merge_sort(nums[mid:])

    return merge(left,right)


def merge(left,right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):    # the loop runs while both list have still elemnts left
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

merge_sort(nums)
```
