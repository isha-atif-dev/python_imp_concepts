# Amortized Time Complexity

- means the average time per operation when an operation sometimes becomes expensive but most of the time its cheap

Side Note: about "dynamic array"

- its the array that automatically grow in size when it becomes full 


```python
# Example append

nums = []
nums.append(1)
nums.append(2)
nums.append(3)
nums

# usually appending to the list is O(1)
```

# But sometimes it becomes O(n)

- because pyhton list are dynamic arrays 

when the array gets full and you append more elments it resizes automatically 

When the List Gets Full

Now we do:

nums.append(5)

Python must:

1️⃣ Create a bigger array (usually double size)

[ _ _ _ _ _ _ _ _ ]

2️⃣ Copy all old elements

[1 2 3 4 _ _ _ _]

3️⃣ Add the new element

[1 2 3 4 5 _ _ _]

This copying step takes:

O(n)

Because all elements must be copied.
This copying takes O(n)


But Why Do We Still Say append() is O(1)?

Because resizing does not happen every time.

Example operations:

append → O(1)

append → O(1)

append → O(1)

append → O(1)

resize → O(n)

append → O(1)

append → O(1)

If we perform many appends, the expensive resizing happens rarely.

So the average cost per operation becomes:

# O(1)

# This is called amortized O(1).

hash table resize is also amortized O(1)
