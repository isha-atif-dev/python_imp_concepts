# imp concept below (mutable trap)


```python
a = [1, 2]
b = a
b = b + [3]

# list is mutable but here 'a' will not chnage why


# Case 1 — Modify in place (mutable behavior)
# b.append(3)

# This changes the SAME object.

# Memory stays same.
# Both a and b see change.


# but 


# Case 2 — Create a new list
# b = b + [3]

# This does NOT modify original list.

# Instead:

# Python creates a NEW list [1, 2, 3]

# Assigns that new list to b

# a still points to original [1, 2]

# So final result:

# a = [1, 2]
# b = [1, 2, 3]
```

```python
# Important Subtlety

# Compare carefully:

# Case 1 — In-place mutation
# b += [3]   # modifies same object
# Case 2 — New object
# b = b + [3]   # creates new list

# That tiny difference is a classic interview trap
```

# summary 

 In Python, variables point to objects in memory. If the object is mutable (like list, dict, set), methods such as append(), extend(), remove(), sort() modify the same object in place. If the object is immutable (like int, str, tuple), any “change” actually creates a new object and rebinds the variable. Operations like +, *, or sorted() typically return a new object rather than mutating the original.
