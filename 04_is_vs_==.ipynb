# ==   (Values equality checking)

- It check do these two object have the same values 

```python
a = [1,2]
b = [1,2]
print(a == b)

# Because values are equal.

# Even though they are different objects in memory.
```

# is  (memory identity checking)
- It check do these two values have the same memory location


```python
# example 1 
num1 = 5
num2 = 5

print(num1 is num2)
```

# Note that example 1 and 2 giving diffenent ans

```python
# example 2 
num1 = 5000
num2 = 5000

print(num1 is num2)
```

- Why these two same examples giving different ans

# Integer Caching

Why?

Python caches small integers (-5 to 256).

This is implementation detail — not something to rely on.

```python
a = [1,2]
b = [1,2]

# both are differnt list 
print(a == b)
print(a is b)   

# Even though both lists are same, they are two different list objects in memory.

# Each [] creates a new list object.
```

```python
x = [1,2]
y = x

print(x == y)
print(x is y)
```
