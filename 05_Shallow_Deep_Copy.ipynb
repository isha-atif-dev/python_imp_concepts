# Shallow Copy

A shallow copy creates:

A new outer container
But inner objects are still shared

```python
a = [[1,2,3],[4,5,6]]

b = a.copy()

b[0].append(6)

b
```

It means that the outer container was copied but inner objects still reference to same memory location

# Deep Copy
Deep copy creates:

A completely independent copy
Including all nested objects

```python
import copy

a = [[1,2,3],[4,5,6]]

b = copy.deepcopy(a)

b[0].append(4)

a
```

Here a still the same but b is another object now pointing to different memory location

# Interview Summary Line

Shallow copy duplicates the outer container but keeps references to nested objects. Deep copy recursively copies all nested objects.

```python
# shallow copy

a = [{"lr": 0.1}, {"lr": 0.01}]
b = a.copy()

b[0]["lr"] = 9

print(a)
```
