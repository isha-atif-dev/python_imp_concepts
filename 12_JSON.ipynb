# JSON (JavaScript Object Notation)

- Its a structured text format for representing data 

{

  "name":"Isha",

  "age":24,

  "is_student":true,
  
  "skills": ["Python","AI"]

}

JSON has **one main job**:

👉 **Send and receive data between systems**

JSON is used whenever systems need to exchange **clean, structured data instead of messy text**

# Examples :


**1. API response (very common)**

When you open a weather app 🌦️

The app calls an API, and it gets data like this (JSON):

- Because the app needs specific values, not sentences.

```python
import json

with open("data.json", "r") as f:
    data = json.load(f)  # converting json text into python dict to be able to use
# it in our app in this way we can easily access speicifc values cuz otherwise
#  the data that we are accessing will come in messy text

print(data["temperature"])

with open("review.json", "w") as f:
    json.dump(data, f, indent=4)
```

json.load()  for accessing file like we are doing here 

json.loads() for accessing string 

⚠️ Small thing to remember

👉 dump = write to file

👉 load = read from file

👉 loads = string → dict

👉 dumps = dict → string
