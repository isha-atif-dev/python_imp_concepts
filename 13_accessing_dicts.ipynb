# Two ways to access dictionary:

1. Strict way (can crash)

data["key"]

2. Safe way (recommended for APIs/AI)

data.get("key")

```python
# for example 

data = {
    "name" : "isha",
    "age" : 23
}

print(data["marks"])  # it gives errors and the app crashes
```

```python
# efficient way to access dicts keys is 

data = {
    "name" : "isha",
    "age" : 23
}

print(data.get("marks","Not avaliable"))
```

When writing code:

**If data is controlled → use []**

**If data is external/AI/user → use .get()**
