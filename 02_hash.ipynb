# Hash

# A fixed-size number generated from an object.

Python uses a hash function hash() to convert data into an integer.


```python
print(hash("hello"))
print(hash(123))
```

```python
# This number is used by dict internally

# but why??

# Because dictionary must be FAST.

```

A hash is an integer generated by Python’s built-in hash() function for a given object. In dictionaries, this hash value is used to determine where a key–value pair should be stored internally. Dictionary keys must be immutable (such as int, str, or tuple with immutable elements) because their hash value must remain stable. If a key could change, its hash would change, and the dictionary would no longer be able to locate it correctly.

When a key is inserted into a dictionary, Python first computes its hash. Since the dictionary has a limited internal table size, Python transforms this large hash number into a valid storage position (index) using an internal calculation (conceptually similar to hash % table_size). The key–value pair is stored at that computed index.

Later, when we access a key, Python recomputes the same hash, derives the same index, and directly jumps to that location in memory instead of scanning the entire dictionary. That is why dictionary lookup is average O(1) time complexity.

```python
# What Is a Dictionary Table?

# A Python dictionary is implemented using a hash table.

# A hash table is basically:

# An array (a list-like structure in memory) with fixed slots.

# Imagine something like this internally:

# Index:   0     1     2     3     4     5     6     7
#         ----  ----  ----  ----  ----  ----  ----  ----
#         None  None  None  None  None  None  None  None

# That array is what I mean by “dictionary table”.

# Each slot can store:

# (key, value)
```

Why Is It Called a “Table”?

Because conceptually it looks like:

Index	Key	    Value
0	    None	
1	    None	
2	    None	
3	    None	
4	    None	
5	    name	Isha
6	    None	
7	    None	

This internal structure is the hash table.

🟢 Important: It Is NOT a 2D Table

It’s just a one-dimensional array in memory.

But each position holds:

(hash, key, value)

Internally Python stores extra metadata too.

```python
# What If Two Keys Want Same Index?

# This is called a hash collision.

# Example:

# hash(key1) % 8 = 3
# hash(key2) % 8 = 3

# Both want slot 3.

# Python then:

# Probes next available slot

# Uses internal collision resolution strategy

# But this is automatic.

# You don’t manage it.
```

# Rehashing 

Python dictionaries do NOT have fixed size forever.They: Automatically resize (grow) when they become too full.

Why Resize?

If too many keys are stored in too few slots:

Collisions increase

Probing increases

Lookup becomes slower

To keep lookup near O(1), Python maintains something called a load factor.

What Is Load Factor?

Load factor ≈

number_of_items / table_size

When this ratio becomes too high (around 2/3 full), Python:

✔ Allocates a bigger table
✔ Recomputes new indices
✔ Re-inserts all existing keys

This is called rehashing.

Does Resize Affect Performance?

Yes — temporarily.

During resizing:

Time complexity becomes O(n)

Because all elements must be reinserted

But resizing does not happen often.

So average complexity remains O(1).

This is called:

Amortized O(1)

Very important interview term.
