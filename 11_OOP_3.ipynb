# Inheritance

- It allows one class to reuse attributes and methods of another class

So instead of rewriting code, a new class can inherit functionality from an existing class.

```python
# Basic syntax of inheritance

class ParentClass:
    pass

class ChildClass(ParentClass):
    pass
```

```python
# Example:

class BaseModel:   # Parent Class
    def __init__(self):
        pass

    def train(self,X,y):
        print("training model")

    def predict(self,X):
        print("predictng the value")

    def evaluate(self):
        pass


class LinearRegression(BaseModel):
    pass

class RandomForest(BaseModel):
    pass

class NeuralNetwork(BaseModel):
    pass


model = LinearRegression()

model.train()
model.predict()
```

Python supports single inheritance and multiple inheritance. In practice, most ML frameworks use single inheritance where specific models inherit common functionality from a base model class.

- Multiple Inheritance: one child has more than one parent
- Hierarchical inheritance: One parent → many children
- Multilevel inheritance: A → B → C chain
- Hybrid inheritance: Combination

# Method Overriding

- Method overriding happens when a child class provides its own implementation of a method that already exists in the parent class.

```python
# Example:

class BaseModel:   # Parent Class
    def __init__(self):
        pass

    def train(self):
        print("Generic training procedure")   #parent class method

    def predict(self):
        print("predictng the value")

    def evaluate(self):
        pass


class LinearRegression(BaseModel):
    def train(self):    # but child class method override that method
        print("Training Linear Regression using least squares")

class RandomForest(BaseModel):
    pass

class NeuralNetwork(BaseModel):
    pass


model = LinearRegression()

print(model.train())
```

# Why overriding is important in ML

Different ML models train differently.

- Model : Training approach
1. Linear Regression: Least squares optimization
2. Neural Network: Gradient descent

But they all share the same interface:

train(X, y)

predict(X)

So we override the training logic.

# Polymorphism

- Polymorphism means the same method name can behave differently depending on the object that calls it.

```python
# for example 
# Imagine we define a base interface for models.

class BaseModel:

    def train(self, X, y):
        pass

    def predict(self, X):
        pass

# Now different models implement these methods differently.

# Linear Regression model
class LinearRegression(BaseModel):

    def train(self, X, y):
        print("Training Linear Regression")

    def predict(self, X):
        print("Predicting using Linear Regression")



# Neural Network
class NeuralNetwork(BaseModel):

    def train(self, X, y):
        print("Training Neural Network using gradient descent")

    def predict(self, X):
        print("Predicting using neural network forward pass")

model1 = LinearRegression()
model2 = NeuralNetwork()

model1.train(X=9, y=9)
model2.train(X=9, y=9)    
```

Same method name:

train()

But different behavior.

That is polymorphism.

- Polymorphism allows the same method name to behave differently depending on the object that calls it. In machine learning libraries, different models implement methods like fit() and predict() differently while sharing the same interface.

```python
class BaseModel:
    def predict(self):
        print("Base prediction")

class RandomForest(BaseModel):
    def predict(self):
        print("Random Forest prediction")

class SVM(BaseModel):
    def predict(self):
        print("SVM prediction")

models = [RandomForest(), SVM()]

for model in models:
    model.predict()
```

# @classmethod and @staticmethod

- These are special types of methods in Python classes. They help organize code better when the method does not depend on a specific object.

the method we already using which depend on object data is Instance method

# Instance Method (normal method)

```python
class Model:

    def train(self, X, y):
        print("training model")

# Here the method receives:

# self

# Meaning:  the specific object calling the method

model = Model()
model.train(X=7, y=7)

# so instance method depend on object state
```

# Class Method

- Sometimes a method should operate on the class itself, not on a specific object.

@classmethod

```python
class Model:

    framework = 'PyTorch'

    @classmethod 
    def get_framework(cls):
        return cls.framework
    
print(Model.get_framework())  # this is how to access class method
```

Notice two things:

1. decorator : @classmethod 
2. cls : cls refers to the class itself.

```python
# Why useful in ML?
# Example: loading a pretrained model.

class Model:

    def __init__(self,name):
        self.name = name

    @classmethod
    def pretrained_model(cls):
        return cls("pretrained-model")      # Model("pretrained-model")  meaning it got ot init fucntion that 
    # has name in it and model_1.name = "pretrained-model"

model_1 = Model.pretrained_model()

print(model_1.name)
```

# Static Method

- A static method is simply a function placed inside a class for logical grouping.

```python
class MathUtils:

    @staticmethod
    def add(a, b):
        return a + b
    
MathUtils.add(2,3)
```

You could write:

def add(a,b):

    return a+b

But sometimes we group functions related to a concept.

Example:

DataUtils.normalize()

DataUtils.standardize()

DataUtils.scale()

So static methods help organize code logically.

# Dunder methods

- “Dunder” means double underscore.

- They are called special methods because Python automatically calls them in certain situations.

```python
# Example:

# __init__
# __repr__
# __str__
# __len__
# __call__
```

# 1️⃣ __repr__ (object representation)

- This controls how an object is displayed.

```python
# without repr the object will look like this

class Model:

    def __init__(self, name):
        self.name = name

    # def __repr__(self):
    #     return f"Model(name={self.name})"
    
model = Model("LinearRegression")
print(model)
```

```python
# with repr
class Model:

    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f"Model(name={self.name})"
    
model = Model("LinearRegression")
print(model)
```

# 2️⃣ __len__  

```python
class Dataset:

    def __init__(self,data):
        self.data = data

    def __len__(self):
        return len(self.data)
    
data = Dataset([1,2,3,4])
len(data)
```

# 3️⃣ __call__

This allows an object to behave like a function.

```python
class SimpleModel:

    def __call__(self,x):
        return x * 2
    
value = SimpleModel()
value(5)
```
