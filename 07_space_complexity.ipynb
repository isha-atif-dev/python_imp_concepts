# Space Complexity

- It describes how much additional memory an algorithm reuqires as the input size increases.

it doesnt measure exact bytes , it measures growth of memory usage

- There are 2 types of memory

1. input Space

memory required to store the input itself.  as it already exists we usually do not count this in space complexity

2. Auxiliary Space (Extra Space)

memory craeted by the algorithm itself.

This counts in Space complexity

# O(1) Constant

```python
# Example 1 : O (1)

def max_num(nums):
    max = nums[0]

    for num in nums:
        if num > max:
            max = num
    
    return max

nums = [1,2,4,2,56,34,23,7]
max_num(nums)


# here the space compexity is constant cuz only 
# max and num var is created which is not dependent on input size 
```

# O(n)

```python
# Example 2 : O(n) linear space


def square(nums):
    result = []

    for num in nums:
        result.append(num * num)
    return result


nums = [1,2,3,4]
square(nums)


# here as we are using result = []  
# which is created addiotnally and also when the input size increased it also increases so 
# O(n)
```

# In-Place Algorithm and Out-of- place Algorithm

# In-Place

- An in place algorithm modifies the original data directly without creating new data strutures

it uses very little extra space

Usually : O(1) Space Complexity

```python
# Example

nums = [1,5,6,2,3]
nums.sort()
print(nums)

# here the same list was modified
# so this is in place
```

# Out of Place

- An out of place create new data sturture to store results

so it uses extra memory

Space Complexity : O(n)

```python
# Example :

nums = [1,4,2,6,8,7,9,0]
sorted_elements = sorted(nums)
sorted_elements
```
