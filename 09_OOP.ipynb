# Object Oriented Programming 

- Object-Oriented Programming is a paradigm that organizes code using classes and objects, where objects encapsulate both data and the methods that operate on that data.

# Classes and objects

1. A class is a blueprint (defines structure + behavior).

2. An object is a runtime instance created from that blueprint.

3. In Python, everything is an object. A class is also an object (created by type).

```python
class MyClass:   # making class
    pass

c1 = MyClass()    # creating instance of that class which is object 
```

```python
class Counter:

    def increment(self):   # Method operates on that object's data.
        self.value += 1

c = Counter()
c.value = 5
c.increment()

print(c.value)
```

# Why self exists

self means:

- It allows the method to work with the data of the object that called it.

Counter.value(c)   this is happening above

```python
class Bank:

    def deposit(self,amount):
        self.balance += amount


acc1 = Bank()
acc1.balance = 1000

acc1.deposit(200)

print(acc1.balance)
```

self is a reference to the current instance of the class, allowing the method to access or modify that object's attributes.

# Initializer __init__

# The problem without __init__

Before learning __init__, look at what we were doing.

class Bank:

    def deposit(self, amount):

        self.balance += amount


acc1 = Bank()

acc1.balance = 1000   # manually attaching attribute

This is bad practice because:

- the object might forget to set balance

- code becomes unsafe

- attributes are not guaranteed to exist

# Solution: __init__

The __init__ method initializes the object’s attributes when an instance is created, ensuring every object starts with a valid state.

__init__ is commonly called the constructor, but technically it is an initializer, not the real constructor.



```python
class Bank_2:

    def __init__(self,balance):
        self.balance = balance

    def deposit(self,amount):
        self.balance += amount

account_1 = Bank_2(20000)
account_1.deposit(300)
print(account_1.balance)

```

Why __init__ is important (interview answer)

It ensures that:

objects start with valid state

required attributes are always created

initialization logic is centralized
