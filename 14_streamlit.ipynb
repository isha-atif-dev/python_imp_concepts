# Streamlit

- Streamlit is a Python library that lets you build web apps using only Python

No need for:

- HTML ❌
- CSS ❌ (optional)
- JavaScript ❌

You just write Python like:

```python
import streamlit as st

st.title("My app")
st.write("Hello world")
```

What kind of apps?

Streamlit is mainly used for:

1. Data science apps
2. ML dashboards
3. AI tools (like YOUR project)
4. quick prototypes

# most important concept

- Streamlit runs your entire script from top to bottom every time something changes

This is called:

👉 RERUN MODEL

```python
if st.button("click me!"):
    st.write("button clicked")

st.text_input("write here!")

# everytime the button is clicked or some changes happnes the streamlit runs again
```

**st.session_state**

It is memory that survives reruns

Without it:
- count = 0
- count += 1

👉 always resets to 0 after rerun

With session_state:


```python
if "count" not in st.session_state:
    st.session_state.count = 0

st.session_state.count += 1

# 👉 value persists
```

# Layout system


```python
col1, col2 = st.columns([5.2, 1.3])

# 👉 splits screen into columns

st.tabs(["Bugs", "Complexity"])

# 👉 creates tabs

# with st.expander("Details"):

# 👉 collapsible section

st.markdown("Hello")

# 👉 It displays formatted text in your app
```

# markdown

- In Streamlit, `st.markdown()` is used to display text, and it can also render HTML and CSS for custom styling.

```python
st.markdown("""
    <div style = "color : red ; font-size : 20px;">
            Styled text</div>
""",unsafe_allow_html=True)

# unsafe_allow_html=True  this will allow htlm to run otherwise html will not run inside markdown
```
